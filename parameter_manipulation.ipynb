```python
from transformers import pipeline
summarization = pipeline('summarization', model='d0rj/rut5-base-summ')
qa = pipeline("question-answering", model="AlexKay/xlm-roberta-large-qa-multilingual-finedtuned-ru")
```

```python
context = '''
Я три года в общаге,
Я в ней ем и в ней сплю,
Я дружу в ней с друзьями
И с девчонкой мучу.
Могут выселить гады,
Комендант у нас зверь.
Каждый день то засада,
То потеря потееерь

Шняга-шняжная, жизнь общажная,
Беспонтовая, но кайфовая.
Жизнь общажная, алкогольная,
Так-то шняжная, но прикольная.

Мы едим здесь пельмени,
С майонезом и без.
В основном с майонезом,
Но бывает и без.
А учёба достала,
Задолбала в конец,
Как сказала бы Алла:
«Это полный пипец!».

Шняга-шняжная, жизнь общажная
Беспонтовая, но кайфовая,
Жизнь общажная, алкогольная,
Так-то шняжная, но прикольная.
'''
```

```python
question = "Какого жить в общаге?"

result = qa(question=question,
            context=context)
answer = result['answer']
print("Ответ-span (по умолчанию):", answer)
```

```python
sentences = context.split('.')
for sentence in sentences:
    if answer in sentence:
        print("Полное предложение с ответом:", sentence.strip())
        break
```

```python
summary = summarization(
    context,
    max_length=60,
    min_length=40,
    do_sample=False
)[0]['summary_text']
print("Обобщение:", summary)
```
